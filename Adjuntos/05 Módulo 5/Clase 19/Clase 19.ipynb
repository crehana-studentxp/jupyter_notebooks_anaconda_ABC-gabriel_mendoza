#### Objetivo: graficar el mes y el importe factura

### Primero se prepara la base de datos

```python
#Obtenemos el año y el mes de la fecha de emisión
df['Año']=df['Fecha de Emision'].dt.year
df['Mes']=df['Fecha de Emision'].dt.month
```

```python
df.head()
```

```python
# Lo segundo será agrupar la base de datos
df=df.groupby(['Año','Mes'],as_index=False)['Importe Factura'].sum()
df
```

```python
df=df[['Mes','Importe Factura']]
df
```

### Gráfico de barras

Una vez que tenemos en una columna el mes y en otra el importe factura podemos generar nuestro primer gráfico

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df.plot(kind='bar', figsize=(10,5))
plt.xlabel('Mes')
plt.ylabel('Importe Factura')
plt.title('Evolución mensual de la facturación')

plt.show()
```
