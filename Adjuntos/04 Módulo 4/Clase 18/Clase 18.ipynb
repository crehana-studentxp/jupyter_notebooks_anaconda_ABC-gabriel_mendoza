```python
# Looping over DataFrames
```

```python
# nos lo da como una serie por lo tanto la tratamos como series
for i in df.iterrows():
    print(i)
```

```python
for i, row in df.iterrows():
    print(row['No. cliente'],row['Vendedor'])
```

#### El problema de iterrows es que lo guarda como una serie y a veces es complicado de manejar: solución itertuples

```python
# lo guardamos en una empty list
lista=[]
```

```python
for row in df.itertuples():
    lista.append(row.Vendedor)
```

```python
lista
```
