### Creamos un Dataframe vacío

### Creamos un Dataframe con diccionarios

```python
empresas= {'Empresas':['Empresa A','Empresa B','Empresa C','Empresa D'],
            'Ventas':[1500000,2000000,1234000,100000],
            'Costos':[600000,543678,235453,21464],
            'Ciudad':['Buenos Aires','Los Angeles','México','Miami'],
            'Piezas':[500,1234,6542,2344]}
```

```python
df=pd.DataFrame(empresas)
df
```

### Creamos un Dataframe con listas

```python
jugadores=[['Lebron','Hoford'],[30,25]]
```

```python
# Indices y columnas
df=pd.DataFrame(jugadores,columns=['Jugadores','Puntos'],index=['Jugadores','Puntos'])
df
```
