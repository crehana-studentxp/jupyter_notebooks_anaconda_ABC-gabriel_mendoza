```python
# Agrupar la base de datos basada en columnas
```

```python
df.groupby('Vendedor').sum()
```

```python
df.groupby('Vendedor')['Importe Factura'].sum()
```

```python
df.groupby('Vendedor')['Importe Factura','Costo'].sum()
```

```python
df.groupby(['Vendedor','Fecha de Pago'])['Importe Factura','Costo'].sum()
```

```python
df.groupby(['Vendedor','Fecha de Pago']).sum()
```

```python
df.groupby(['Vendedor','Fecha de Pago']).mean()
```

```python
# Para quitar el índice
df.groupby(['Vendedor','Fecha de Pago'],as_index=False).mean()
```

```python
df.groupby('Vendedor').mean()
```

```python
df.groupby('Vendedor',as_index=False).mean()
```

### Agg method

```python
df.groupby('Vendedor',as_index=False)['Importe Pagado','% Comision Vendedor'].agg({'Importe Pagado':'sum','% Comision Vendedor':'mean'})
```
